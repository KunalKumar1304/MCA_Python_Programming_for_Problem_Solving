```python
import math

def math_operations():
    # a) Square root of a number
    num_sqrt = float(input("Enter a number to find its square root: "))
    if num_sqrt < 0:
        print("Error: Cannot calculate the square root of a negative number.")
    else:
        print(f"Square root of {num_sqrt} is: {math.sqrt(num_sqrt)}")
    # b) Factorial of a number
    num_fact = int(input("\nEnter a non-negative integer for factorial: "))
    if num_fact < 0:
        print("Error: Factorial is not defined for negative numbers.")
    else:
        print(f"Factorial of {num_fact} is: {math.factorial(num_fact)}")
    # c) Power of a number (base^exponent)
    base = float(input("\nEnter the base number: "))
    exponent = float(input("Enter the exponent: "))
    print(f"{base} raised to the power of {exponent} is: {math.pow(base, exponent)}")

if __name__ == "__main__":
    math_operations()
```
