```python
import random

# List of students
students = ["Aman", "Mukesh", "Kunal", "Pushkar", "Lokesh", "Jagdish","Leeladhar"]

# Select CR randomly
cr = random.choice(students)
print("Selected CR:", cr)

# Shuffle the student list
random.shuffle(students)

# Print shuffled list
print("Shuffled Student List:", students)
```
