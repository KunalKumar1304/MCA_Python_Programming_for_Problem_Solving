```python
class Point:
    """Represents a point in 2D space"""

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def display(self):
        return f"({self.x}, {self.y})"


class Circle:
    """Represents a circle using a Point as center"""

    def __init__(self, center, radius):
        self.center = center      # Point object
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def circumference(self):
        return 2 * 3.14159 * self.radius

    def display(self):
        return f"Center: {self.center.display()}, Radius: {self.radius}"


# -------- Demonstration --------
if __name__ == "__main__":
    p = Point(2, 3)
    c = Circle(p, 5)

    print(c.display())
    print("Area:", c.area())
    print("Circumference:", c.circumference())
```
